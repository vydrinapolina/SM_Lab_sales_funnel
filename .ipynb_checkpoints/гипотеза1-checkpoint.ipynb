**Основная гипотеза (H₁):**  
Товары, которые доступны (product_is_now / product_is_delivery / product_is_pick_up / product_is_pick_point), имеют более высокие конверсии из просмотра в корзину и из корзины в заказ.  
**Альтернативная (H₀):**  
Наличие флагов доступности не влияет на вероятность добавления в корзину или оформления заказа.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_parquet('sm_data_opt.parquet')
```

```python
# общий признак доступности
df['is_available'] = (df['product_is_pick_up'] | df['product_is_pick_point'] | df['product_is_delivery'] | df['product_is_now'])
```

```python
availability_analysis = df.groupby('is_available').agg({
    'event_id': 'count',
    'add_to_cart_flag': 'sum',
    'create_order_flag': 'sum'
}).reset_index()

# конверсии
availability_analysis['view_to_cart_%'] = (availability_analysis['add_to_cart_flag'] / availability_analysis['event_id']) * 100
availability_analysis['cart_to_order_%'] = (availability_analysis['create_order_flag'] / availability_analysis['add_to_cart_flag']) * 100
availability_analysis['view_to_order_%'] = (availability_analysis['create_order_flag'] / availability_analysis['event_id']) * 100

availability_analysis
```

```python
# График для сравнения конверсий
metrics_to_plot = ['view_to_cart_%', 'cart_to_order_%', 'view_to_order_%']
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

for i, metric in enumerate(metrics_to_plot):
    data = availability_analysis[['is_available', metric]]
    labels = ['Недоступен', 'Доступен']
    values = [data[data['is_available'] == False][metric].values[0], 
              data[data['is_available'] == True][metric].values[0]]
    
    axes[i].bar(labels, values, color=['red', 'green'], alpha=0.7)
    axes[i].set_title(f'Конверсия: {metric}')
    axes[i].set_ylabel('Проценты (%)')
    
    # Добавляем значения на столбцы
    for j, v in enumerate(values):
        axes[i].text(j, v + 0.1, f'{v:.1f}%', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

### Общий Z-test

```python
from statsmodels.stats.proportion import proportions_ztest

# доступные товары
available_views = availability_analysis[availability_analysis['is_available'] == True]['event_id'].values[0]
available_carts = availability_analysis[availability_analysis['is_available'] == True]['add_to_cart_flag'].values[0]
available_orders = availability_analysis[availability_analysis['is_available'] == True]['create_order_flag'].values[0]

# недоступные товары
unavailable_views = availability_analysis[availability_analysis['is_available'] == False]['event_id'].values[0]
unavailable_carts = availability_analysis[availability_analysis['is_available'] == False]['add_to_cart_flag'].values[0]
unavailable_orders = availability_analysis[availability_analysis['is_available'] == False]['create_order_flag'].values[0]

# конверсия в корзину
count_cart = [available_carts, unavailable_carts]
nobs_cart = [available_views, unavailable_views]
z_stat_cart, p_value_cart = proportions_ztest(count_cart, nobs_cart)

# конверсия из корзины в заказ
count_order = [available_orders, unavailable_orders]
nobs_order = [available_carts, unavailable_carts]
z_stat_order, p_value_order = proportions_ztest(count_order, nobs_order)

print(f"Статистическая значимость различий:")
print(f"конверсия в корзину: z-statistic = {z_stat_cart:.4f}, p-value = {p_value_cart:.4f}")
print(f"конверсия из корзины в заказ: z-statistic = {z_stat_order:.4f}, p-value = {p_value_order:.4f}")
```

Почему конверсия из корзины в заказ выше у недоступных товаров?

### Z-test по устройствам

```python
df_dev = device_conversion_analysis.copy()
results = []
devices = df_dev['device_type'].unique()

for device in devices:
    subset = df_dev[df_dev['device_type'] == device]

    row_avail = subset[subset['is_available'] == True].iloc[0]
    row_not_avail = subset[subset['is_available'] == False].iloc[0]

    # конверсия в корзину
    successes_vc = np.array([row_avail['add_to_cart_flag'], row_not_avail['add_to_cart_flag']])
    observations_vc = np.array([row_avail['event_id'], row_not_avail['event_id']])
    z_vc, p_vc = proportions_ztest(count=successes_vc, nobs=observations_vc)

    # конверсия из корзины в заказ
    successes_co = np.array([row_avail['create_order_flag'], row_not_avail['create_order_flag']])   
    observations_co = np.array([row_avail['add_to_cart_flag'], row_not_avail['add_to_cart_flag']])
    z_co, p_co = proportions_ztest(count=successes_co, nobs=observations_co)

    results.append({
        'device_type': device,
        'z_view_to_cart': z_vc,
        'p_view_to_cart': p_vc,
        'z_cart_to_order': z_co,
        'p_cart_to_order': p_co
    })

device_z_results = pd.DataFrame(results)
device_z_results
```

Обратим внимание, что на этапе добавления товара в корзину, наличие любого из флагов доступности играет важную роль. То есть гипотезу H1 можно частично подтвердить (но после стат. теста).  
Общая интерпретация по cart → order: Гипотеза НЕ подтверждается и эффект разный по категориям. Так, в категориях одежда / обувь недоступные товары лучше конвертируются.  

Получается такой эффект: недоступные товары получают другой тип трафика  
→ Меньше просмотров, но более мотивированные пользователи  
→ Конверсия считается на меньшей, более “тёплой” выборке

```python
df_brand = brand_analysis.copy()
results = []
brands = df_brand['brand_lvl'].unique()

for brand in brands:
    subset = df_brand[df_brand['brand_lvl'] == brand]

    row_avail = subset[subset['is_available'] == True].iloc[0]
    row_not_avail = subset[subset['is_available'] == False].iloc[0]

    # в корзину
    successes_vc = np.array([row_avail['carts'], row_not_avail['carts']])
    observations_vc = np.array([row_avail['views'], row_not_avail['views']])
    z_vc, p_vc = proportions_ztest(successes_vc, observations_vc)

    # в заказ
    successes_co = np.array([row_avail['orders'], row_not_avail['orders']])
    observations_co = np.array([row_avail['carts'], row_not_avail['carts']])

    z_co, p_co = proportions_ztest(successes_co, observations_co)
    results.append({
        'brand_lvl': brand,
        'z_view_to_cart': z_vc,
        'p_view_to_cart': p_vc,
        'z_cart_to_order': z_co,
        'p_cart_to_order': p_co
    })

brand_z_results = pd.DataFrame(results)
brand_z_results
```

### Авторизация, мобильное приложение и премиальные бренды. Как совокупность этих факторов влияет на конверсию в заказ

```python
# Премиальность: объединяем уровни 0 и 1 в один признак
df['is_premium'] = df['brand_lvl'].isin([0, 1])
# Моб. приложение: объединяем Android и iOS
df['is_mobile_app'] = df['device_type'].isin(['mobileapp_android', 'mobileapp_ios'])
# Категория: объединяем одежда и обувь
df['is_clothing_shoes'] = df['product_category'].isin(['одежда', 'обувь'])
```

```python
filtered_df = df[
    (df['is_premium'] == True) &
    (df['is_clothing_shoes'] == True) &
    (df['is_mobile_app'] == True)
]
```

```python
analysis_grouped = filtered_df.groupby(['is_available', 'user_is_authorized'], observed=True).agg({
        'event_id': 'count',
        'add_to_cart_flag': 'sum',
        'create_order_flag': 'sum'
}).reset_index()
```

```python
# Расчет конверсий
analysis_grouped['view_to_cart_%'] = (analysis_grouped['add_to_cart_flag'] / analysis_grouped['event_id']) * 100
analysis_grouped['cart_to_order_%'] = (analysis_grouped['create_order_flag'] / analysis_grouped['add_to_cart_flag'].replace(0, 1)) * 100

analysis_grouped['is_available_str'] = analysis_grouped['is_available'].apply(lambda x: 'Доступен' if x == True else 'Недоступен' if x == False else str(x))
analysis_grouped['user_is_authorized_str'] = analysis_grouped['user_is_authorized'].apply(lambda x: 'Авторизован' if x == True else 'Не авторизован' if x == False else str(x))

print(analysis_grouped[['is_available_str', 'user_is_authorized_str', 'view_to_cart_%', 'cart_to_order_%']].to_string(index=False))
```

### Z-test по авторизации в разрезе действий пользователей с премиальной одеждой и обувью в мобильных приложениях

Тест 1: Сравнивает конверсию в заказ **авторизованных** и **неавторизованных** пользователей для **недоступных товаров**  
Тест 2: Сравнивает конверсию в заказ **авторизованных** и **неавторизованных** пользователей для **доступных товаров**  
Тест 3: Сравнивает конверсию в заказ **авторизованных** пользователей между доступными и недоступными товарами  
Тест 4: Сравнивает конверсию в заказ **неавторизованных** пользователей между доступными и недоступными товарами

```python
data_dict = {}
for idx, row in analysis_grouped.iterrows():
    key = (row['is_available_str'], row['user_is_authorized_str'])
    data_dict[key] = {
        'event_id': row['event_id'],
        'add_to_cart': row['add_to_cart_flag'],
        'create_order': row['create_order_flag'],
        'view_to_cart_%': row['view_to_cart_%'],
        'cart_to_order_%': row['cart_to_order_%']
    }
```

```python
# Тест 1. Сравнение конверсии в заказ для НЕДОСТУПНЫХ товаров

# Группа A: Авторизованные, недоступные товары
# Группа B: Неавторизованные, недоступные товары
group_a = data_dict[('Недоступен', 'Авторизован')]
group_b = data_dict[('Недоступен', 'Не авторизован')]

success_a = group_a['create_order']
total_a = group_a['add_to_cart']
success_b = group_b['create_order']
total_b = group_b['add_to_cart']

if total_a > 0 and total_b > 0:
    count = [success_a, success_b]
    nobs = [total_a, total_b]
    z_stat, p_value = proportions_ztest(count, nobs, alternative='two-sided')
    print(f"z-statistic = {z_stat:.4f}, p-value = {p_value:.4f}")
    
    if p_value < 0.05:
        print("Различие СТАТИСТИЧЕСКИ ЗНАЧИМО")
        print("Авторизованные пользователи действительно чаще доводят до заказа недоступные товары")
    else:
        print("Различие статистически НЕ значимо")
```

```python
# Тест 2. Сравнение конверсии в заказ для ДОСТУПНЫХ товаров

group_c = data_dict[('Доступен', 'Авторизован')]
group_d = data_dict[('Доступен', 'Не авторизован')]

success_c = group_c['create_order']
total_c = group_c['add_to_cart']
success_d = group_d['create_order']
total_d = group_d['add_to_cart']

if total_c > 0 and total_d > 0:
    count = [success_c, success_d]
    nobs = [total_c, total_d]
    z_stat, p_value = proportions_ztest(count, nobs, alternative='two-sided')
    print(f"z-statistic = {z_stat:.4f}, p-value = {p_value:.4f}")
    
    if p_value < 0.05:
        print("Различие СТАТИСТИЧЕСКИ ЗНАЧИМО")
        print("Авторизованные пользователи чаще доводят до заказа доступные товары")
    else:
        print("Различие НЕ статистически значимо")
```

```python
# Тест 3: Сравнение авторизованных пользователей между доступными и недоступными товарами

group_e = data_dict[('Недоступен', 'Авторизован')]
group_f = data_dict[('Доступен', 'Авторизован')]

success_e = group_e['create_order']
total_e = group_e['add_to_cart']
success_f = group_f['create_order']
total_f = group_f['add_to_cart']

if total_e > 0 and total_f > 0:
    count = [success_e, success_f]
    nobs = [total_e, total_f]
    z_stat, p_value = proportions_ztest(count, nobs, alternative='two-sided')
    print(f"z-statistic = {z_stat:.4f}, p-value = {p_value:.4f}")
    
    if p_value < 0.05:
        print("Различие СТАТИСТИЧЕСКИ ЗНАЧИМО")
        print("Авторизованные пользователи чаще оформляют заказ на недоступные товары")
    else:
        print("Различие НЕ статистически значимо")
```

```python
# Тест 4. Сравнение неавторизованных пользователей между доступными и недоступными товарами

group_g = data_dict[('Недоступен', 'Не авторизован')]
group_h = data_dict[('Доступен', 'Не авторизован')]

success_g = group_g['create_order']
total_g = group_g['add_to_cart']
success_h = group_h['create_order']
total_h = group_h['add_to_cart']

if total_g > 0 and total_h > 0:
    count = [success_g, success_h]
    nobs = [total_g, total_h]
    z_stat, p_value = proportions_ztest(count, nobs, alternative='two-sided')
    print(f"z-statistic = {z_stat:.4f}, p-value = {p_value:.4f}")
    
    if p_value < 0.05:
        print("Различие СТАТИСТИЧЕСКИ ЗНАЧИМО (p < 0.05)")
        print("Неавторизованные пользователи имеют РАЗНУЮ конверсию для доступных и недоступных товаров")
    else:
        print("Различие статистически НЕ значимо (p ≥ 0.05)")
```

### Результаты тестов:
**Тест 1**: Сравнивает конверсию в заказ авторизованных и неавторизованных пользователей для недоступных товаров. Различие **статистически значимо**. Авторизованные пользователи действительно чаще доводят до заказа недоступные товары.  
**Тест 2**: Сравнивает конверсию в заказ авторизованных и неавторизованных пользователей для доступных товаров. Различие **статистически значимо**. Авторизованные пользователи доводят до заказа доступные товары чаще, чем неавторизованные.  
**Тест 3**: Сравнивает конверсию в заказ авторизованных пользователей между доступными и недоступными товарами. Различие **статистически значимо**. Авторизованные пользователи чаще оформляют заказ на недоступные товары.  
**Тест 4**: Сравнивает конверсию в заказ неавторизованных пользователей между доступными и недоступными товарами. Разница между конверсией в заказ доступных товаров и конверсией в заказ недоступных товаров неавторизованных пользователей **статистически не значимо**.
