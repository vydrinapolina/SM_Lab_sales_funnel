```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_parquet('sm_data_opt.parquet')
```

### Анализ товаров по категориям

```python
category = df.groupby('product_category', observed=True).agg({'product_id': 'nunique'}).reset_index()
category.columns = ['product_category', 'unique_products_count']

gender = df.groupby(['product_category', 'product_gender'], observed=True).agg({'product_id': 'nunique'}).unstack(fill_value=0)
gender.columns = ['unique_female_products', 'unique_male_products', 'unique_unisex_products']

brand = df.groupby(['product_category', 'brand_lvl'], observed=True).agg({'product_id': 'nunique'}).unstack(fill_value=0)
brand.columns = ['unique_lux_products', 'unique_mid_products', 'unique_low_products']

category_merged = category.merge(
    gender, 
    left_on='product_category', 
    right_index=True, 
    how='left'
).merge(
    brand, 
    left_on='product_category', 
    right_index=True, 
    how='left'
)

category_merged
```

```python
plt.figure(figsize=(15, 5))

# распределение товаров по категориям
plt.subplot(1, 3, 1)
plt.pie(category_merged['unique_products_count'], 
        labels=category_merged['product_category'],
       autopct='%1.1f%%')
plt.title('Распределение товаров по категориям')

# распределение по полу
plt.subplot(1, 3, 2)
gender_totals = category_merged[['unique_female_products', 'unique_male_products', 'unique_unisex_products']].sum()
plt.pie(gender_totals, 
        labels=['Женские', 'Мужские', 'Унисекс'], 
        autopct='%1.1f%%')
plt.title('Распределение товаров по полу')

# распределение по уровню брендов
plt.subplot(1, 3, 3)
brand_totals = category_merged[['unique_lux_products', 'unique_mid_products', 'unique_low_products']].sum()
plt.pie(brand_totals, 
        labels=['Премиум', 'Средние', 'Бюджетные'], 
        autopct='%1.1f%%')
plt.title('Распределение по уровню брендов')

plt.tight_layout()
plt.show()
```

```python
# Активность по категориям товаров
category_activity = df.groupby('product_category', observed=True).agg({
    'event_id': 'count',  # просмотры
    'add_to_cart_flag': 'sum',  # добавления в корзину
    'create_order_flag': 'sum'  # заказы
}).rename(columns={
    'event_id': 'total_views',
    'add_to_cart_flag': 'total_carts', 
    'create_order_flag': 'total_orders'
}).sort_values('total_views', ascending=False)

category_activity
```

```python
plt.figure(figsize=(15, 5))

# просмотры
plt.subplot(1, 3, 1)
plt.pie(category_activity['total_views'], 
        labels=category_activity.index,
       autopct='%1.1f%%')
plt.title('Распределение просмотров по категориям')

# добавления в корзину
plt.subplot(1, 3, 2)
plt.pie(category_activity['total_carts'], 
        labels=category_activity.index, 
        autopct='%1.1f%%')
plt.title('Распределение добавлений в корзину по категориям')

# оформление заказа
plt.subplot(1, 3, 3)
plt.pie(category_activity['total_orders'], 
        labels=category_activity.index, 
        autopct='%1.1f%%')
plt.title('Распределение оформлений заказов по категориям')

plt.tight_layout()
plt.show()
```

### Анализ товаров по брендам

```python
brand_segment_basic = df.groupby('brand_lvl', observed=True).agg({
    'event_id': 'count',
    'add_to_cart_flag': 'sum',
    'create_order_flag': 'sum',
    'product_id': 'nunique'
}).reset_index()

brand_discounted = df[df['ret_discount'] > 0].groupby('brand_lvl', observed=True).agg({
    'product_id': 'nunique'
}).reset_index()
brand_discounted.columns = ['brand_lvl', 'discounted_products']

brand_gender_unique = df.groupby(['brand_lvl', 'product_gender'], observed=True).agg({
    'product_id': 'nunique'
}).unstack(fill_value=0)
brand_gender_unique.columns = ['female_products', 'male_products', 'unisex_products']
brand_gender_unique = brand_gender_unique.reset_index()

brand_segment = brand_segment_basic.merge(brand_discounted, on='brand_lvl')
brand_segment = brand_segment.merge(brand_gender_unique, on='brand_lvl')

brand_segment.columns = [
    'brand_segment',
    'views',
    'cart_adds', 
    'orders',
    'unique_products',
    'discounted_products',
    'female_products',
    'male_products',
    'unisex_products'
]

brand_segment
```

```python
plt.figure(figsize=(15, 6))

# активности
plt.subplot(1, 2, 1)
x = np.arange(len(brand_segment))
width = 0.25

plt.bar(x - width, brand_segment['views'], width, label='Просмотры', color='skyblue', alpha=0.8)
plt.bar(x, brand_segment['cart_adds'], width, label='Добавления в корзину', color='lightcoral', alpha=0.8)
plt.bar(x + width, brand_segment['orders'], width, label='Заказы', color='lightgreen', alpha=0.8)

plt.xlabel('Сегменты брендов')
plt.ylabel('Количество')
plt.title('Активность пользователей по сегментам')
plt.xticks(x, ['Премиум', 'Средний', 'Бюджет'])
plt.legend()

for i, v in enumerate(brand_segment['views']):
    plt.text(i - width, v + 100000, f'{v/1000000:.1f}M', ha='center', va='bottom', fontsize=8)
for i, v in enumerate(brand_segment['cart_adds']):
    plt.text(i, v + 100000, f'{v/1000:.0f}K', ha='center', va='bottom', fontsize=8)
for i, v in enumerate(brand_segment['orders']):
    plt.text(i + width, v + 100000, f'{v/1000:.0f}K', ha='center', va='bottom', fontsize=8)

# конверсии
plt.subplot(1, 2, 2)

conversion_to_cart = (brand_segment['cart_adds'] / brand_segment['views'] * 100).round(1)
conversion_to_order = (brand_segment['orders'] / brand_segment['cart_adds'] * 100).round(1)

x_conv = np.arange(len(brand_segment))
plt.bar(x_conv - 0.2, conversion_to_cart, 0.4, label='В корзину', color='coral', alpha=0.8)
plt.bar(x_conv + 0.2, conversion_to_order, 0.4, label='В заказ', color='mediumseagreen', alpha=0.8)

plt.xlabel('Сегменты брендов')
plt.ylabel('Конверсия (%)')
plt.title('Конверсия между этапами воронки')
plt.xticks(x_conv, ['Премиум', 'Средний', 'Бюджет'])
plt.legend()

for i, (conv_cart, conv_order) in enumerate(zip(conversion_to_cart, conversion_to_order)):
    plt.text(i - 0.2, conv_cart + 1, f'{conv_cart}%', ha='center', va='bottom', fontsize=10, fontweight='bold')
    plt.text(i + 0.2, conv_order + 1, f'{conv_order}%', ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 5))

# Распределение товаров по сегментам
plt.subplot(1, 3, 1)
plt.pie(brand_segment['unique_products'], 
        labels=['Премиум', 'Средний', 'Бюджет'], colors = ['lightyellow', 'lightgray', 'lightgreen'],
        autopct='%1.1f%%')
plt.title('Распределение уникальных товаров по сегментам')

# Товары со скидкой
plt.subplot(1, 3, 2)
plt.pie(brand_segment['discounted_products'], 
        labels=['Премиум', 'Средний', 'Бюджет'], 
        autopct='%1.1f%%')
plt.title('Распределение товаров со скидкой')

# Распределение товаров по полу
plt.subplot(1, 3, 3)
segments = ['Премиум', 'Средний', 'Бюджет']
female = brand_segment['female_products']
male = brand_segment['male_products']
unisex = brand_segment['unisex_products']

x = range(len(segments))
plt.bar(x, female, label='Женские', color='pink')
plt.bar(x, male, bottom=female, label='Мужские', color='lightblue')
plt.bar(x, unisex, bottom=[i+j for i,j in zip(female, male)], label='Унисекс', color='lightgreen')

plt.xticks(x, segments)
plt.title('Распределение товаров по полу')
plt.legend()

plt.tight_layout()
plt.show()
```
